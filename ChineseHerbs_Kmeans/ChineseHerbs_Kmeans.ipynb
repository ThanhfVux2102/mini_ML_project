```python
import os
from google.colab.patches import cv2_imshow
import cv2
import matplotlib.pyplot as plt
from google.colab import drive
drive.mount('/content/gdrive')
img = cv2.imread('/content/gdrive/MyDrive/herb_ds/herb_ds.jpg')


# default color-space of cv2 is BGR, SO WE HAVE TO REVERT TO RGB
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img);
```

docs for input data requirements of k-means clustering : https://medium.com/@evgen.ryzhkov/5-stages-of-data-preprocessing-for-k-means-clustering-b755426f9932

```python

import numpy as np


pixel_values =  img.reshape(-1,3)

# float because just numerical ( rules )
pixel_values = np.float32(pixel_values)



```

```python
from sklearn.cluster import KMeans
k = 2
kmeans = KMeans(n_clusters=k, n_init=10, random_state=42)
labels = kmeans.fit_predict(pixel_values)
centers = np.uint8(kmeans.cluster_centers_)


segmented = centers[labels.flatten()]
segmented_image = segmented.reshape(img.shape)
plt.imshow(segmented_image)
plt.title("After K-means Clustering (k = 2)")
plt.axis("off")
```
